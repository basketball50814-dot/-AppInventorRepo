### **2. [APCS 實戰技巧]：一行讀取多個數字**

在 ZeroJudge 題目中，常要求在一行內讀入兩個數字（如：`5 10`），我們使用 `split()` 切開字串，並用 `map()` 轉成整數。


#### **[示範區]：使用 `map()` 處理一行三個數字的輸入**

你可以很容易地將 `map()` 和 `split()` 的組合擴展到處理三個或更多數字的情況。

```
# 示範：輸入三個數字並計算總和與平均值
print("請輸入三個數字，用空白鍵隔開 (例如: 10 20 30)：")

# 讀取使用者輸入的三個數字，使用 split() 分割，map(int, ...) 轉換為整數
num1, num2, num3 = map(int, input().split())

# 計算總和
sum_three = num1 + num2 + num3

# 計算平均值 (結果為浮點數)
average_three = sum_three / 3

# 印出輸入的數字
print(f"您輸入的三個數字是 {num1}, {num2}, 和 {num3}")

# 印出總和
print(f"它們的總和是 {sum_three}")

# 印出平均值，並保留兩位小數
print(f"它們的平均值是 {average_three:.2f}")
```
